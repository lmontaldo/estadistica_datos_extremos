```python
import pyextremes
import yfinance as yf
import pandas as pd

# Download historical data for a specific stock (e.g., Apple Inc.)
stock_data = yf.download('AAPL', period="max")#Apple Inc

# Calculate daily returns
stock_data['Daily Return'] = stock_data['Adj Close'].pct_change()

# Drop NaN values
stock_data = stock_data.dropna()
stock_data.to_csv(path+'/AAPL_daily_returns.csv')
print(stock_data['Daily Return'])
```

```python
# Define the ticker symbol for the S&P 500
ticker_symbol = "^GSPC"

# Fetch historical data for the S&P 500
sp500 = yf.Ticker(ticker_symbol)

# Get historical market data
histsp500 = sp500.history(period="max")  # 'max' for all available data; you can customize the period
histsp500.to_csv(path+'/sp500_Closing_values.csv')
# Display the first few rows of the dataframe
print(histsp500.head())
```

```python
import matplotlib.pyplot as plt

# Plot the closing prices
histsp500['Close'].plot(title="S&P 500 Historical Closing Values")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```

```python
# Define the ticker symbol for WTI Crude Oil
ticker_symbol = "CL=F"

# Fetch historical data for WTI Crude Oil
crude_oil = yf.Ticker(ticker_symbol)

# Get historical market data
histcrude_oil = crude_oil.history(period="max")  # 'max' for all available data; customize as needed
histcrude_oil.to_csv(path+'/wti_Closing_values.csv')
# Display the first few rows of the dataframe
print(histcrude_oil.head())
```

```python
# Plot the closing prices
histcrude_oil['Close'].plot(title="WTI Crude Oil Historical Closing Prices")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.show()
```
